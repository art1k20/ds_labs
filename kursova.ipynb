```python
from mlxtend.preprocessing import minmax_scaling
plt.figure(figsize=(20,50))
scaled_data = df.drop(['Other names','ISO 3166-1 alpha-3 CODE', 'Continent','Death percentage','Population'], axis = 1)
scaled_data = minmax_scaling(scaled_data, columns= [ 'Total Cases', 'Total Deaths', 'Tot Cases//1M pop', 'Tot Deaths/1M pop'])
sns.heatmap(data=scaled_data, annot= True)
```

```python

plt.figure(figsize=(10,20))
sns.histplot(data= scaled_data)
```

```python
scaled_data_100 = scaled_data[:100]

plt.figure(figsize=(30,10))

plot = sns.lineplot(scaled_data_100) 

plt.xticks(rotation = 90)
for i in [ 'Total Cases', 'Total Deaths', 'Tot Cases//1M pop', 'Tot Deaths/1M pop']:

   

    avrg = scaled_data_100.mean(axis=0)[i]
    for j in range(2, len(scaled_data_100[i]+1)):
        if round(scaled_data_100[i][j], 2) == round(avrg, 2):     
            x = j
    print(avrg, x)
    plot.annotate(str(avrg), xy=(x, avrg), horizontalalignment='center')
```

```python
fig, axes = plt.subplots(3, 3, sharex = True, figsize = (40, 20))
Ukraine  = scaled_data.loc['Ukraine']
USA = scaled_data.loc['USA']
Poland = scaled_data.loc['Poland']
Norway = scaled_data.loc['Norway']
Lithuania = scaled_data.loc['Lithuania']
Japan = scaled_data.loc['Japan']
Estonia = scaled_data.loc['Estonia']
Argentina = scaled_data.loc['Argentina']
Australia = scaled_data.loc['Australia']
Ukr_plot = sns.barplot(ax = axes[0,0], x = Ukraine.values, y = Ukraine.index)
Ukr_plot.set_title('Ukraine')
USA_plot = sns.barplot(ax = axes[0, 1], x = USA.values, y = USA.index)
USA_plot.set_title('USA')
Pol_plot = sns.barplot(ax = axes[0,2], x = Poland.values, y = Poland.index)
Pol_plot.set_title('Poland')
Nor_plot = sns.barplot(ax = axes[1,0], x = Norway.values, y = Norway.index)
Nor_plot.set_title('Norway')
Lith_plot = sns.barplot(ax = axes[1,1], x = Lithuania.values, y = Lithuania.index)
Lith_plot.set_title('Lithuania')
Jap_plot = sns.barplot(ax = axes[1,2], x = Japan.values, y = Japan.index)
Jap_plot.set_title('Japan')
Est_plot = sns.barplot(ax = axes[2,0], x = Estonia.values, y = Estonia.index)
Est_plot.set_title('Estonia')
Arg_plot =  sns.barplot(ax = axes[2,1], x = Argentina.values, y = Argentina.index)
Arg_plot.set_title('Argentina')
Aust_plot = sns.barplot(ax = axes[2,2], x = Australia.values, y = Australia.index)
Aust_plot.set_title('Australia')
```
