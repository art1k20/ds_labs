```python
scaled_data = df.drop(['Other names','ISO 3166-1 alpha-3 CODE', 'Continent','Death percentage','Population'], axis = 1)
plt.figure(figsize=(30,10))
sns.barplot(x = scaled_data.index[:100] , y = scaled_data['Total Cases'][:100])
plt.xticks(rotation = 90)
```

```python
from mlxtend.preprocessing import minmax_scaling
plt.figure(figsize=(20,50))
scaled_data = df.drop(['Other names','ISO 3166-1 alpha-3 CODE', 'Continent','Death percentage','Population'], axis = 1)
scaled_data = minmax_scaling(scaled_data, columns= [ 'Total Cases', 'Total Deaths', 'Tot Cases//1M pop', 'Tot Deaths/1M pop'])
sns.heatmap(data=scaled_data, annot= True)
```
